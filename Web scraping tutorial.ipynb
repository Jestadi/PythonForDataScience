```python
with open('07_float_align.html','r') as html_file:
    content = html_file.read()
    print(content)
```

```python
soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

```python
tags = soup.find_all('h3')
for tag in tags:
    print(tag.text)
```

```python
course_cards = soup.find_all('div', class_='box')
for course in course_cards:
    course_name = course.h3.text
    print(course_name)
```
